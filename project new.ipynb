## DATA PRE-PROCESSING

```python

# CONVERT TO DATE FORMAT

df['Time Period Start Date'] = pd.to_datetime(df['Time Period Start Date'])
df['Time Period End Date'] = pd.to_datetime(df['Time Period End Date'])
print(df['Time Period Start Date'])
print(df['Time Period End Date'])
print(df[['Time Period Start Date','Time Period End Date']].dtypes)
```

```python
# CHECK MISSING VALUES AND DUPLICATES

print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
# HANDLING MISSING VALUES

df["Value"] = df["Value"].fillna(df["Value"].mean())
print (df["Value"])
df["LowCI"] = df["LowCI"].fillna(df["LowCI"].mean())
print (df["LowCI"])
df["HighCI"] = df["HighCI"].fillna(df["HighCI"].mean())
print (df["HighCI"])
df["Suppression Flag"] = df["Suppression Flag"].fillna(df["Suppression Flag"].mean())
print (df["Suppression Flag"])
df["Confidence Interval"] = df["Confidence Interval"].fillna(df["Confidence Interval"].mode().iloc[0])
print (df["Confidence Interval"])
df["Quartile Range"] = df["Quartile Range"].fillna(df["Quartile Range"].mode().iloc[0])
print (df["Quartile Range"])
```

```python
# AFTER HANDILING MISSING VALUES

print(df.isnull().sum())
```

```python
# DERIVED NEW COLUMNS

df['Month'] = df['Time Period Start Date'].dt.to_period('M').astype(str)
print(df['Month'])
df['Value_category'] = pd.qcut(df['Value'], 4, labels=['Low','Medium-Low','Medium-High','High'])
print(df['Value_category'])
```

The tail beyond 40–60 suggests outliers or exceptional cases.
